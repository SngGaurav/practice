# A. Pizza Metrics
1. How many pizzas were ordered?
2. How many unique customer orders were made?
3. How many successful orders were delivered by each runner?
4. How many of each type of pizza was delivered?
5. How many Vegetarian and Meatlovers were ordered by each customer?
6. What was the maximum number of pizzas delivered in a single order?
7. For each customer, how many delivered pizzas had at least 1 change and how many had no changes?
8. How many pizzas were delivered that had both exclusions and extras?
9. What was the total volume of pizzas ordered for each hour of the day?
10. What was the volume of orders for each day of the week?

```python
%%sql
# How many pizzas were ordered?
SELECT 
  COUNT(PIZZA_ID) AS TOTAL_ORDERS 
FROM 
  CUSTOMER_ORDERS
```

```python
%%sql
# How many unique customer orders were made?
SELECT 
  COUNT(DISTINCT CUSTOMER_ID) AS TOTAL_CUSTOMERS 
FROM 
  CUSTOMER_ORDERS
```

```python
%%sql
# How many successful orders were delivered by each runner?
SELECT 
  RUNNER_ID, 
  SUM(
    CASE WHEN CANCELLATION IN ('null', '') 
    OR CANCELLATION IS NULL then 1 ELSE 0 END
  ) AS SUCCESSFUL_ORDERS 
FROM 
  RUNNER_ORDERS 
GROUP BY 
  RUNNER_ID
```

```python
%%sql
# How many of each type of pizza was delivered?
SELECT 
  PIZZA_ID, 
  COUNT(CO.ORDER_ID) AS TOTAL_DELIVERED 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
WHERE 
  CANCELLATION IN ('null', '') 
  OR CANCELLATION IS NULL 
GROUP BY 
  PIZZA_ID
```

```python
%%sql
# How many Vegetarian and Meatlovers were ordered by each customer?
SELECT 
  CUSTOMER_ID, 
  PIZZA_NAME, 
  COUNT(CO.PIZZA_ID) AS TOTAL_ORDERED 
FROM 
  CUSTOMER_ORDERS CO 
  INNER JOIN PIZZA_NAMES PN ON CO.PIZZA_ID = PN.PIZZA_ID 
GROUP BY 
  CUSTOMER_ID, 
  PIZZA_NAME
```

```python
%%sql
# What was the maximum number of pizzas delivered in a single order?

SELECT 
  COUNT(CO.PIZZA_ID) AS MAX_DELIVERED 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
WHERE 
  CANCELLATION IN ('null', '') 
  OR CANCELLATION IS NULL 
GROUP BY 
  CO.ORDER_ID 
ORDER BY 
  MAX_DELIVERED DESC 
LIMIT 
  1
```

```python
%%sql
# For each customer, how many delivered pizzas had at least 1 change and how many had no changes?

SELECT 
  CO.CUSTOMER_ID, 
  SUM(
    CASE WHEN (
      exclusions IN ('null', '') 
      OR exclusions IS NULL
    ) 
    AND (
      extras IN ('null', '') 
      OR extras IS NULL
    ) THEN 1 ELSE 0 END
  ) AS NO_CHANGE, 
  SUM(
    CASE WHEN (
      exclusions IN ('null', '') 
      OR exclusions IS NULL
    ) 
    AND (
      extras IN ('null', '') 
      OR extras IS NULL
    ) THEN 0 ELSE 1 END
  ) AS CHANGES 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
WHERE 
  CANCELLATION IN ('null', '') 
  OR CANCELLATION IS NULL 
GROUP BY 
  CO.CUSTOMER_ID
```

```python
%%sql
# How many pizzas were delivered that had both exclusions and extras?

SELECT 
  COUNT(1) AS CHANGES_PIZZAS 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
WHERE 
  (
    CANCELLATION IN ('null', '') 
    OR CANCELLATION IS NULL
  ) 
  AND (
    exclusions NOT IN ('null', '') 
    AND exclusions IS NOT NULL
  ) 
  AND (
    extras NOT IN ('null', '') 
    AND extras IS NOT NULL
  )
```

```python
%%sql
# What was the total volume of pizzas ordered for each hour of the day?

SELECT 
  STRFTIME("%H", ORDER_TIME) AS ORDER_HOUR, 
  COUNT(1) as NUM_PIZZA 
FROM 
  CUSTOMER_ORDERS 
GROUP BY 
  ORDER_HOUR
```

```python
%%sql
# What was the volume of orders for each day of the week?

WITH TEMP_TBL AS (
  SELECT 
    CAST(
      STRFTIME("%w", ORDER_TIME) AS INTEGER
    ) AS ORDER_WEEK_DAY, 
    COUNT(1) as NUM_PIZZA 
  FROM 
    CUSTOMER_ORDERS 
  GROUP BY 
    ORDER_WEEK_DAY
) 
SELECT
  CASE ORDER_WEEK_DAY
    WHEN 0 THEN 'Sunday'
    WHEN 1 THEN 'Monday'
    WHEN 2 THEN 'Tuesday'
    WHEN 3 THEN 'Wednesday'
    WHEN 4 THEN 'Thursday'
    WHEN 5 THEN 'Friday'
    WHEN 6 THEN 'Saturday'
    END AS day_of_week,
    NUM_PIZZA
FROM
  TEMP_TBL
ORDER
  BY NUM_PIZZA
```

```python
%%sql
# How many runners signed up for each 1 week period? (i.e. week starts 2021-01-01)

SELECT 
  STRFTIME("%W", REGISTRATION_DATE) AS REGISTRATION_WEEK, 
  COUNT(1) AS TOTAL_RUNNERS 
FROM 
  RUNNERS 
GROUP BY 
  REGISTRATION_WEEK
```

```python
%%sql
# What was the average time in minutes it took for each runner to arrive at the Pizza Runner HQ to pickup the order?

SELECT 
  RUNNER_ID, 
  AVG(
    (
      (
        julianday(PICKUP_TIME) - julianday(order_time)
      )* 24 * 60
    )
  ) AVG_TIME 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
GROUP BY 
  RUNNER_ID
```

```python
%%sql
# Is there any relationship between the number of pizzas and how long the order takes to prepare?

WITH TEMP_TBL AS (
  SELECT 
    co.order_id, 
    count(pizza_id) AS NUM_OF_PIZZA, 
    (
      (
        julianday(PICKUP_TIME) - julianday(order_time)
      )* 24 * 60
    ) TOTAL_TIME 
  FROM 
    CUSTOMER_ORDERS CO 
    JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
  WHERE 
    (
      CANCELLATION IN ('null', '') 
      OR CANCELLATION IS NULL
    ) 
  GROUP BY 
    co.order_id
) 
SELECT 
  NUM_OF_PIZZA, 
  ROUND(
    AVG(TOTAL_TIME), 
    2
  ) AS AVG_PREPARITION_TIME 
FROM 
  TEMP_TBL 
GROUP BY 
  NUM_OF_PIZZA
```

```python
%%sql
# What was the average distance travelled for each customer?

SELECT 
  CUSTOMER_ID, 
  ROUND(
    AVG(
      CAST(
        CASE WHEN distance LIKE '%km%' THEN TRIM(DISTANCE, 'km') ELSE DISTANCE END AS FLOAT
      )
    ), 
    2
  ) AS Distnace_in_km 
FROM 
  CUSTOMER_ORDERS CO 
  JOIN RUNNER_ORDERS RO ON CO.ORDER_ID = RO.ORDER_ID 
WHERE 
  (
    CANCELLATION IN ('null', '') 
    OR CANCELLATION IS NULL
  ) 
GROUP BY 
  CUSTOMER_ID
```

```python
%%sql
# What was the difference between the longest and shortest delivery times for all orders?

SELECT 
  MAX(DURATION) - MIN(DURATION) TOTAL_TIME 
FROM 
  RUNNER_ORDERS 
WHERE 
  (
    CANCELLATION IN ('null', '') 
    OR CANCELLATION IS NULL
  )
```

```python
%%sql
# What was the average speed for each runner for each delivery and do you notice any trend for these values?

WITH TEMP_TBL AS (
  SELECT 
    *, 
    CAST(
      CASE WHEN distance LIKE '%km%' THEN TRIM(DISTANCE, 'km') ELSE DISTANCE END AS FLOAT
    ) AS DISTANCE_KM, 
    CAST(
      CASE WHEN DURATION LIKE '%MIN%' THEN TRIM(DURATION, '%MIN%') ELSE DURATION END AS INTEGER
    ) AS DURATION_MIN 
  FROM 
    RUNNER_ORDERS
) 
SELECT 
  runner_id, 
  order_id, 
  PICKUP_TIME, 
  ROUND(
    AVG(
      distance_km / (
        CAST(duration_min AS REAL)/ 60
      )
    ), 
    1
  ) AS avg_speed_kmph, 
  ROUND(
    AVG(
      (distance_km / 1.609) / (
        CAST(duration_min AS REAL) / 60
      )
    ), 
    1
  ) AS avg_speed_mph 
FROM 
  TEMP_TBL 
WHERE 
  (
    CANCELLATION IN ('null', '') 
    OR CANCELLATION IS NULL
  ) 
GROUP BY 
  runner_id, 
  order_id, 
  PICKUP_TIME
```

```python
%%sql
# What is the successful delivery percentage for each runner?

SELECT 
  runner_id, 
  SUM(
    CASE WHEN (
      CANCELLATION IN ('null', '') 
      OR CANCELLATION IS NULL
    ) THEN 1 ELSE 0 END
  ) / CAST(
    COUNT(order_id) AS REAL
  ) * 100 AS success_pct 
FROM 
  RUNNER_ORDERS 
GROUP BY 
  runner_id
```

```python
%%sql
# What are the standard ingredients for each pizza?

WITH RECURSIVE recipes (pizza_id, topping_id, rest) AS (
    SELECT PIZZA_ID, '', toppings || ',' FROM pizza_recipes WHERE PIZZA_ID
    UNION ALL
      SELECT 
          pizza_id,
         substr(rest, 0, instr(rest, ',')),
        substr(rest, instr(rest, ',')+1)
      FROM recipes
      WHERE rest <> ''
    )
SELECT p.pizza_id, pizza_name, topping_name
FROM pizza_names p
inner join recipes r
    on p.pizza_id = r.pizza_id
inner join pizza_toppings t
    on r.topping_id = t.topping_id
WHERE r.topping_id <> ''
ORDER BY pizza_name;
```

```python
%%sql
# What was the most commonly added extra?

WITH RECURSIVE extras_tbl (content, topping_id, rest) AS (
select '', 
        '', 
        extras || ',' --comma(,) is required at end, else it stuck in loop
        from customer_orders where extras is not Null and extras not in ('', 'null')
union all
select '', substr(rest, 0, instr(rest, ',')),
            substr(rest, instr(rest, ',')+1)
            from extras_tbl where rest <> ''
)
select topping_id, count(1) as added_count from extras_tbl
where topping_id <> ''
group by topping_id
order by added_count desc
```

```python
%%sql
# What was the most common exclusion?

WITH RECURSIVE exclusions_tbl (content, topping_id, rest) AS (
select '', 
        '', 
        exclusions || ',' --comma(,) is required at end, else it stuck in loop
        from customer_orders where exclusions is not Null and exclusions not in ('', 'null')
union all
select '', substr(rest, 0, instr(rest, ',')),
            substr(rest, instr(rest, ',')+1)
            from exclusions_tbl where rest <> ''
)
select topping_id, count(1) as added_count from exclusions_tbl
where topping_id <> ''
group by topping_id
order by added_count desc
```

```python
%%sql
# Generate an order item for each record in the customers_orders table in the format of one of the following:
# 1. Meat Lovers
# 2. Meat Lovers - Exclude Beef
# 3. Meat Lovers - Extra Bacon
# 4. Meat Lovers - Exclude Cheese, Bacon - Extra Mushroom, Peppers

WITH RECURSIVE exclusions_tbl (content, topping_id, rest) AS (
  select 
    order_id || '-' || pizza_id || '-' || ifnull(exclusions, '')|| '-' || ifnull(extras, ''), 
    --if we try adding null value to string, the string become null, no error will come
    '', 
    exclusions || ',' --comma(,) is required at end, else it stuck in loop
  from 
    customer_orders -- where exclusions is not Null and exclusions not in ('', 'null')
  union all 
  select 
    content, 
    substr(
      rest, 
      0, 
      instr(rest, ',')
    ), 
    substr(
      rest, 
      instr(rest, ',')+ 1
    ) 
  from 
    exclusions_tbl 
  where 
    rest <> ''
), 
extras_tbl (content, topping_id, rest) AS (
  select 
    order_id || '-' || pizza_id || '-' || ifnull(exclusions, '')|| '-' || ifnull(extras, ''), 
    '', 
    extras || ',' --comma(,) is required at end, else it stuck in loop
  from 
    customer_orders 
  union all 
  select 
    content, 
    substr(
      rest, 
      0, 
      instr(rest, ',')
    ), 
    substr(
      rest, 
      instr(rest, ',')+ 1
    ) 
  from 
    extras_tbl 
  where 
    rest <> ''
), 
extras_tbl1 as (
  select 
    e.content, 
    group_concat(
      ifnull(topping_name, ''), 
      ', '
    ) as extras 
  from 
    extras_tbl e 
    left join pizza_toppings pt on pt.topping_id = e.topping_id 
  group by 
    e.content
), 
exclusions_tbl1 as (
  select 
    e.content, 
    group_concat(
      ifnull(topping_name, ''), 
      ', '
    ) as exclusions 
  from 
    exclusions_tbl e 
    left join pizza_toppings pt on pt.topping_id = e.topping_id 
  group by 
    e.content
) 
, 
final_tbl as (
  select 
    e.content, 
    ltrim(extras, ', ') as extras, 
    ltrim(exclusions, ', ') as exclusions, 
    substr(
      e.content, 
      0, 
      instr(e.content, '-')
    ) as order_id, 
    substr(
      substr(
        e.content, 
        instr(e.content, '-')+ 1
      ), 
      0, 
      instr(
        substr(
          e.content, 
          instr(e.content, '-')+ 1
        ), 
        '-'
      )
    ) as pizza_id 
  from 
    extras_tbl1 e 
    inner join exclusions_tbl1 ex on e.content = ex.content --where e.content <> Null and ex.content <> Null
    ) 
select 
  order_id, 
  case when (extras <> '') and (exclusions <> '')
        then pizza_name || ' - ' || 'Extra ' || extras || ' - ' || 'Exclude ' || exclusions
      when (extras <> '') and (exclusions = '') 
        then pizza_name || ' - ' || 'Extra ' || extras
      when (extras = '') and (exclusions <> '') 
        then pizza_name || ' - ' || 'Exclude ' || exclusions
      else pizza_name 
  end as order_item 
from 
  final_tbl as f 
  inner join pizza_names as p 
    on f.pizza_id = p.pizza_id
```

```python
%%sql
# Generate an alphabetically ordered comma separated ingredient list for each pizza order 
# from the customer_orders table and add a 2x in front of any relevant ingredients
# For example: "Meat Lovers: 2xBacon, Beef, ... , Salami"

with removed_tbl as (
    select order_id,
        extras, 
        case when instr(exclusions, ',') <> 0 then replace(
        replace(
            toppings, 
            substr(exclusions,
                   0, 
                   instr(exclusions, ',')), 
            ''),
        substr(exclusions,
               instr(exclusions, ',')+1),
         ' ')
            when exclusions is not null and exclusions not in ('null', '') then replace(toppings, exclusions, '')
            else toppings end as removed
    from customer_orders c 
    inner join pizza_recipes p 
        on p.pizza_id = c.pizza_id
),
    added_tbl as (
    select order_id, 
        case when extras is not null and extras not in ('null', '') then removed || ', ' || extras
            else removed end as toppings
    from removed_tbl
    ),
cleaned_tbl as (
select order_id, ltrim(replace(toppings, ', ,', ','), ',') as toppings
    from added_tbl
),
exploded_tbl (order_id, topping_id, rest) as (
    select order_id, '', toppings || ' ,' from cleaned_tbl
    union all
    SELECT order_id,
            substr(rest, 0, instr(rest, ',')),
            substr(rest, instr(rest, ',')+1)
          FROM exploded_tbl
          WHERE rest <> ''
), final_tbl as (
select order_id, topping_name, sum(1) as total from exploded_tbl et
inner join pizza_toppings as pt
    on pt.topping_id = et.topping_id
where et.topping_id <> ''
group by order_id, topping_name
order by order_id, topping_name)

select order_id, 
    group_concat(case when total > 1 then total || 'x' || topping_name
        else topping_name end, ', ') as order_item
from final_tbl
group by order_id
```

```python
%%sql
# What is the total quantity of each ingredient used in all delivered pizzas sorted by most frequent first?

with removed_tbl as (
    select c.order_id,
        extras, 
        case when instr(exclusions, ',') <> 0 then replace(
        replace(
            toppings, 
            substr(exclusions,
                   0, 
                   instr(exclusions, ',')), 
            ''),
        substr(exclusions,
               instr(exclusions, ',')+1),
         ' ')
            when exclusions is not null and exclusions not in ('null', '') then replace(toppings, exclusions, '')
            else toppings end as removed
    from customer_orders c 
    inner join pizza_recipes p 
        on p.pizza_id = c.pizza_id
    inner join runner_orders ro
        on ro.order_id = c.order_id
    where ro.cancellation is Null or ro.cancellation in ('null', '')
),
    added_tbl as (
    select order_id, 
        case when extras is not null and extras not in ('null', '') then removed || ', ' || extras
            else removed end as toppings
    from removed_tbl
    ),
cleaned_tbl as (
select order_id, ltrim(replace(toppings, ', ,', ','), ',') as toppings
    from added_tbl
),
exploded_tbl (order_id, topping_id, rest) as (
    select order_id, '', toppings || ' ,' from cleaned_tbl
    union all
    SELECT order_id,
            substr(rest, 0, instr(rest, ',')),
            substr(rest, instr(rest, ',')+1)
          FROM exploded_tbl
          WHERE rest <> ''
), final_tbl as (
select order_id, topping_name, sum(1) as total from exploded_tbl et
inner join pizza_toppings as pt
    on pt.topping_id = et.topping_id
where et.topping_id <> ''
group by order_id, topping_name
order by order_id, topping_name)

select topping_name, count(1) as Total_Quantity
from final_tbl
group by topping_name
order by Total_Quantity desc
```
